# Example of usage

```python
from funcode.funcode import solve
from funcode.integrators import *
from funcode.odes import harmonic_oscillator

import numpy as np
import matplotlib.pyplot as plt
```

## Problem definition

```python
OMEGA_0 = 1
ZETA = 0.5
system = harmonic_oscillator(OMEGA_0, ZETA)

y_0 = np.r_[1.0, 
            0.0]

t = np.linspace(0, 2*np.pi, 51)
```

## Problem resolution

```python
y_fe = solve(system, y_0, t, forward_euler)
y_pc = solve(system, y_0, t, predcorr_euler)
```

## Plot and compare results

```python
[fig, ax] = plt.subplots()
ax.set_xlabel('Time')
ax.set_ylabel('State')
ax.plot(t, y_fe, '--', label='Euler')
ax.plot(t, y_pc, label='PredCorr')
ax.legend()

plt.show()
```
